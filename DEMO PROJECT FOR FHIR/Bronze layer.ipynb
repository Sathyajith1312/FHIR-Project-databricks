```python
dbutils.fs.ls("/Volumes/sathyajith/demoprojectfhir/rawdata")
```

```python
raw_df=spark.read.option("multiLine","true").json("/Volumes/sathyajith/demoprojectfhir/rawdata/*.json")
```

```python
from pyspark.sql.functions import current_timestamp

bronze_df = (
    raw_df
    .withColumn("ingestion_time", current_timestamp())
    .withColumn("source_file", raw_df["_metadata.file_path"])
)
```

```python
%sql
CREATE TABLE IF NOT EXISTS sathyajith.demoprojectfhir.bronze_fhir_bundle
USING DELTA
```

```python
bronze_df.write \
    .format("delta") \
    .option("mergeSchema", "true") \
    .mode("append") \
    .saveAsTable(
      "sathyajith.demoprojectfhir.bronze_fhir_bundle"
    )
```

```python
%sql
SELECT COUNT(*) FROM sathyajith.demoprojectfhir.bronze_fhir_bundle;
```

```python
%sql
SELECT
  COUNT(*) AS total_bundles
FROM sathyajith.demoprojectfhir.bronze_fhir_bundle;
```
